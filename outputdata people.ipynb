```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Load input data
df = pd.read_csv("output.csv")
```

```python
df.head()
```

```python
# Sort data by employee identifiers and effective dates
df.sort_values(by=['Employee Code', 'Effective Date'], inplace=True)
```

```python
df["End Date"].fillna("2100-01-01", inplace=True)
df["Manager Employee Code"].fillna(1.0, inplace=True)
df["Last Pay Raise Date"].fillna("2023-01-01", inplace=True)
df["Variable Pay"].fillna(0, inplace=True)
df["Tenure in Org"].fillna(0, inplace=True)
```

```python
df.head(20)
```

```python
df["Last Compensation"].fillna(df["Last Compensation"].mean(), inplace=True)
df["Performance Rating"].fillna(df["Performance Rating"].mean(), inplace=True)
df["Engagement Score"].fillna(df["Engagement Score"].mean(), inplace=True)
```

```python
df.head()
```

```python
# Initialize historical records DataFrame
historical_df = pd.DataFrame(columns=df.columns)
```

```python
# Iterate through each employee
for employee_id, employee_data in df.groupby('Effective Date'):
    previous_end_date = None
    
# Iterate through each record for the employee
    for index, record in employee_data.iterrows():
        effective_date = record['Effective Date']
        
        # Assuming effective_date is a string in the format 'YYYY-MM-DD'
        effective_date_str = record['Effective Date']
        effective_date = datetime.strptime(effective_date_str, '%Y-%m-%d')
        
        # Determine end date to avoid overlap
        end_date = effective_date - timedelta(days=1) if previous_end_date else datetime(2100, 1, 1)
        
        # Create historical record
        historical_record = record.copy()
        historical_record['Effective Date'] = effective_date
        historical_record['End Date'] = end_date
        
        # Append record to historical_df
        historical_df = historical_df.append(historical_record, ignore_index=True)
        
        # Update previous_end_date for the next iteration
        previous_end_date = end_date
```

```python
historical_df.head()
```

```python
# Save transformed data to a new CSV file
output_file = "output_historical.csv"
historical_df.to_csv(output_file, index=False)
```
